```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
import zarr  # to read in Zarr format
import gcsfs # to read google cloud data
import momlevel as ml
import cftime
import cmocean
```

```python
xr.set_options(display_style='html')
%matplotlib inline                           
%config InlineBackend.figure_format='retina' 
plt.rcParams['figure.figsize'] = 12,6  
```

```python
gcs = gcsfs.GCSFileSystem(token='anon')
```

```python
google_cloud_cmip6 = pd.read_csv('https://storage.googleapis.com/cmip6/cmip6-zarr-consolidated-stores.csv')
```

```python
google_cloud_cmip6
```

# <span style="color:red"> Data: Atmosphere monthly water vapor content (prw) on a regular grid (gr) from GFDL-ESM4 from the SSP1-26 climate scenarioe

```python
gfdl_esm4_prw126_data = google_cloud_cmip6.query(
    "activity_id=='ScenarioMIP' & \
    table_id == 'Amon' & \
    variable_id == 'prw' & \
    experiment_id == 'ssp126' & \
    institution_id == 'NOAA-GFDL' &  \
    source_id == 'GFDL-ESM4' & \
    grid_label == 'gr1' & \
    member_id == 'r1i1p1f1'"
)

gfdl_esm4_prw126_data
```

```python
zstore_esm4_prw126 = gfdl_esm4_prw126_data.zstore.values[-1]
mapper_esm4_prw126 = gcs.get_mapper(zstore_esm4_prw126)
gfdl_esm4_prw126 = xr.open_zarr(mapper_esm4_prw126, 
                                consolidated=True,use_cftime=True).prw.rename({'lat':'latitude',
                                                             'lon':'longitude'})
gfdl_esm4_prw126
```

```python
gfdl_esm4_prw126.sel(time=slice('2023-12-01')).mean(dim='time').plot()
plt.title('Water vapor in presence')
```

```python
esm4_wv_15_25 = gfdl_esm4_prw126.sel(time=slice('2015-01-16','2025-12-16')).mean(dim='time')
## do the same thing below but for the last 20 years.
esm4_wv_81_100 = gfdl_esm4_prw126.sel(time=slice('2081-01-16','2100-12-16')).mean(dim='time')

difference_esm4 = esm4_wv_81_100 - esm4_wv_15_25
difference_esm4
```

```python
difference_esm4.plot()
```

```python
difference_esm4.sel(longitude=336,method='nearest').plot.contourf(yincrease=False,
                                                         vmin=-2, vmax=2,
                                                         cmap=cmocean.cm.balance, levels=50)
```

```python
gfdl_esm4_prw126.mean(dim='time').plot()
plt.title('Average water vapor')
```

```python
gfdl_esm4_prw126.sel(time=slice('2100-12-16')).mean(dim='time').plot()
plt.title('Water vapor in "2100"')
```
